```python
import pandas as pd
import requests
import lxml.html as lh
from requests import get
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as bs

links_list=[]
for i in range(1,18):
    url = "https://neu.peopleadmin.com/postings/search?commit=Search&page="+str(i)+"&utf8=%E2%9C%93"
    response = get(url)
    html_soup = bs(response.text,'html.parser')## python's in built library HTML parser
    all_jobs = html_soup.select('#search_results')
    #print(all_jobs)
    all_jobnames = html_soup.select('.col-md-4 a')
    for value in  all_jobnames:
        links_list.append( 'https://neu.peopleadmin.com'+value['href'])
       
    
    for link in links_list:
        response = get(link)
        html_soup1 = bs(response.text,'html.parser')
        print(link)
        
        page = requests.get(link)
        #Store the contents of the website under doc
        doc = lh.fromstring(page.content)
        #Declaring the file name in which the data will be stored in CSV format 
        filename = "NortheasternJobs.csv"
        f = open(filename, "w")
        #Headers for the column names of the CSV file
        headers = "Position Title,Requisition Number,Division/College,Interdisciplinary Division/College,Location,Full-time/Part-time,Grade,Posting Date,Responsibilities,Qualifications,Additional Information \n"
        f.write(headers)

        #Code starts
        td_elements = doc.xpath('//td')
        #Create empty list
        col=[]
        i=0
        #For each row, store each first element (header) and an empty list
        for t in td_elements:
            i+=1
            name=(t.text_content()+',').replace(",", " ")
            print (name)
            print('----------------------')
            col.append((name,[]))
            f.write(name.replace("\n", "")+ "," )
    
    f.close()

        
        
```
